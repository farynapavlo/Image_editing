```
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```
shift = 30
PATH = '../content/sunset-in-vancouver.jpg'
```

```
def show_hist(image: Image):
    image = image.convert('L')
    plt.plot(list(range(256)), image.histogram())
    plt.show()
```

```
img = Image.open(PATH)
img.load()
show_hist(img)
```

```
plt.imshow(img)
```

```
data = np.array(img)
```

```
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        for k in range(3):
            pixel = data[i][j][k] - shift
            if pixel < 0:
                data[i][j][k] = 0
            else:
                data[i][j][k] = pixel
```

```
img = Image.fromarray(data, 'RGB')
show_hist(img)
```

```
plt.imshow(img)
```
